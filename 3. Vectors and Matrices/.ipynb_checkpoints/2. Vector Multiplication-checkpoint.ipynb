```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

v = np.array([2, 1])
w = 2 *v
print (w)

# Plot w
origin = [0], [0]
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, *w, scale=10)
plt.show()
```

```python
b = v / 2
print(b)

# Plot b
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, *b, scale=10)
plt.show()
```

```python
import numpy as np
v = np.array([2,1])
s = np.array([-3,2])
d = np.dot(v,s)
print(d)
```

```python
#in python 3.5 and later, use the @ to calculate the dot product
import numpy as np
v = np.array([2,1])
s = np.array([-3, 2])
d = v @ s
print (d)
```

```python
import math
import numpy as np

#define our vectors
v = np.array([2,1])
s = np.array([-3, 2])

#get the magnitudes
vMag = np.linalg.norm(v)
sMag = np.linalg.norm(s)

#calculate the cosine of theta
cos = np.dot(v,s)/(vMag * sMag)

#so Theta (in degrees) is:
theta = math.degrees(math.acos(cos))
print(theta)
```
